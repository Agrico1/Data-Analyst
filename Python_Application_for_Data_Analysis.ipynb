## Pengelolaan Data `Date`

```python
#7
#Mengubah tipe kolom Date menjadi Datetime
df['order_date']= pd.to_datetime(df['order_date'])
df['ship_date']= pd.to_datetime(df['ship_date'])
df.dtypes
```

```python
#8
#Membuat kolom tambahan untuk day, month, month_num
df['day']=df['order_date'].dt.day_name()
df['month']=df['order_date'].dt.month_name()
df['month_num']=df['order_date'].dt.month
df['year']=df['order_date'].dt.year
df['ship_delay'] = df['ship_date']-df['order_date']
df['ship_delay'] = df['ship_delay'].dt.days
df.head(5)
```

```python
#9
#Latihan: Filter berdasarkan bulan October tahun 2015
# Cara coba sendiri
# filter_data = df[(df['month'] == 'October') & (df['year'] == 2015)]
# filter_data

# cara yang bener
df[(df["order_date"].dt.month ==10)&(df["order_date"].dt.year ==2015)]
```

##Penggunaan fungsi `pivot_table` & `groupby`

### `pivot_table`

**Formula:**

nama_variable  =  `pd.pivot_table`(nama_dataframe`, values=`['kolom_nilai']`,index=`['baris']`,columns=`['header']`,aggfunc=np.`jenis_agregasi)

Output: `DataFrame`

```python
#10
df_pivot = pd.pivot_table(df,values=['quantity'],index=['region'],columns=['category'],aggfunc=np.sum)
df_pivot
```

```python
#11
#Buatlah tabel pivot untuk: index = country, header/kolom = city, nilainya = penjumlahan dari profit
df_pivot_tes= pd.pivot_table(df,values=['profit'],index=['country'],columns=['city'],aggfunc=np.sum)
df_pivot_tes
```

### `groupby`

**Formula:**

nama variabel = nama_dataframe`.groupby(by=`['kolom_1'  ,   'kolom_2'])['kolom_agregasi']`.`jenis_agregasi()

Output: `series`


```python
#12
df_groupby = df.groupby(by=["month_num","month"])["profit"].sum()
df_groupby
```

```python
#13
#merubah output groupby menjadi DataFrame dengan as_index = False
df_groupby = df.groupby(by=["month_num","month"], as_index=False)["profit"].sum()
df_groupby
```

```python
df_groupby = df.groupby(by="city", as_index=False)["profit"].sum()
df_top5 = df_groupby.nlargest(5, "profit")
df_top5
```

```python
# Lakukan operasi groupby pada kolom "city" dan kemudian sum pada kolom "profit"
df_groupby = df.groupby(by="city", as_index=False)["profit"].sum()

# Urutkan DataFrame berdasarkan kolom "profit" secara descending
df_sorted = df_groupby.sort_values(by="profit", ascending=False)

# Ambil 5 baris pertama setelah pengurutan menggunakan head()
df_top5 = df_sorted.head(5)

# Tampilkan DataFrame hasil top 5 city dengan profit terbesar
(df_top5)

```

##Membuat dan menggunakan fungsi sederhana pada DataFrame

Formula:

`def` nama_fungsi `(`nilai`):`

    kondisi



```python
#15
#Membuat suatu fungsi
def condition(value):
    if value > 2:
      return 'High'
    elif value == 2:
      return 'Medium'
    else:
        return 'Low'
```

```python
#16
#Pengaplikasian fungsi
df['status'] = df['quantity'].apply(condition)
df.head()
```

```python
#17
#Buatlah suatu kolom baru yang bernama ship_status yang dapat menyatakan bahwa apabila waktu pengiriman <= 4 hari dari tanggal order maka 'Good' selain itu maka 'Bad'
#Membuat suatu fungsi
def kondisi_ship(value):
    if value <= 4:
      return 'Good'
    else:
        return 'Bad'
```

```python
df['ship_status'] = df['ship_delay'].apply(kondisi_ship)
df.head()
```

## Membuat visualisasi secara sederhana dengan menggunakan `plot`

```python
#18
#Membuat grafik
df_groupby.plot(x='month',
           y=['profit'],
           kind='bar',
           grid = True,
           xlabel = 'Month',
           ylabel = 'Total Profit',
           figsize=(12,7),
           rot = 0, #Rotasi
           title = 'Monthly Total Quantity',
           table = False,
          #  sort_columns = False,
           secondary_y = False)
```

## Menjalankan SQL di Colab dengan `sqlite3`

```python
#19
#Menjalankan SQL di Colab
from sqlite3 import connect
conn = connect(':memory:')
df.to_sql('tokopaedi',conn, index=False, if_exists='replace')
```

```python
#20
#Query SQL
df_sql = pd.read_sql("""
SELECT
    tokopaedi.city,
    sum(tokopaedi.profit) profit
from
    tokopaedi
group by
    1
order by
    2 desc
limit 10
""", conn)
df_sql
```
